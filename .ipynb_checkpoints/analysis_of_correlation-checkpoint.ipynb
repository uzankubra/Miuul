```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)
df= pd.read_csv("cancer.csv")
```

```python
df= df.iloc[:, 1: -1]
```

```python
df.head()
```

```python
cat_cols= [col for col in df.columns if str(df[col].dtypes) in ["category", "bool", "object"]]
```

```python
num_but_cat= [col for col in df.columns if df[col].nunique()<10 and df[col].dtypes in ["int", "float"]]
```

```python
cat_but_car=[col for col in df.columns if df[col].nunique()>20 and str(df[col].dtypes) in ["category", "object"]]
```

```python
cat_cols=[col for col in cat_cols if col not in cat_but_car]
```

```python
cat_cols= cat_cols + num_but_cat
```

```python
num_cols=[col for col in df.columns if df[col].dtypes in ["int", "float"]]
```

```python
num_cols= [col for col in num_cols if col not in cat_cols]
```

```python
corr=df[num_cols].corr()
```

```python
corr
```

```python
sns.set(rc={'figure.figsize': (12,12)})
sns.heatmap(corr, cmap="RdBu")
plt.show()
```

## Yuksek korelasyonlu degiskenlerin silinmesi 

```python
cor_matrix= df.corr().abs()
```

```python
upper_iangle_matrix= cor_matrix.where(np.triu(np.ones(cor_matrix.shape), k=1).astype(np.bool))
```

```python
upper_iangle_matrix
```

```python
drop_list=[col for col in upper_iangle_matrix.columns if any(upper_iangle_matrix[col]>0.90)]
```

```python
drop_list
```

```python
cor_matrix[drop_list]
```

```python
df.drop(drop_list, axis=1)
```

```python
def high_correlated_cols(dataframe, plot=False, corr_th=0.90):
    corr= dataframe.corr()
    cor_matrix=corr.abs()
    upper_iangle_matrix= cor_matrix.where(np.triu(np.ones(cor_matrix.shape), k=1).astype(np.bool))
    drop_list=[col for col in upper_iangle_matrix.columns if any(upper_iangle_matrix[col]>0.90)]
    if plot:
        import seaborn as sns
        import matplotlib.pyplot as plt
        sns.set(rc={'figure.figsize': (12,12)})
        sns.heatmap(corr, cmap="RdBu")
        plt.show()
    return drop_list
```

```python
high_correlated_cols(df)
```

```python
high_correlated_cols(df.drop(drop_list, axis=1), plot=True)
```
